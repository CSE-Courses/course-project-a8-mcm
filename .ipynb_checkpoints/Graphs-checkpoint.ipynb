```python
# Import webscraped data 
from webscrape import *

# Importing numpy and panda packages
import numpy as np
import pandas as pd

# Data Visualization
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from datetime import datetime

# Training and testing package
from sklearn.model_selection import train_test_split

# Linear Regression Model
from sklearn.linear_model import LinearRegression

# Allows plot to appear directly in notebook
%matplotlib inline
```

```python
# Asking user for stock they want to look up
url = get_url("aapl")

# Write the data into csv form and gives it as output.csv
write_to_csv(url)
```

```python
# Read data into a dataframe using panda
df = pd.read_csv(r'output.csv')
# Reversed the data to have the oldest date and data at the top and newest at the bottom
df = df.reindex(index=df.index[::-1])
```

```python
df.plot(x="Date", y="Open", figsize=(16,8), rot=90)
df.plot(x="Date", y="High", figsize=(16,8), rot=90)
df.plot(x="Date", y="Low", figsize=(16,8), rot=90)
df.plot(x="Date", y="Close*", figsize=(16,8), rot=90)
df.plot(x="Date", y="Adj Close**", figsize=(16,8), rot=90)
```
